# **Basic Operations : Resizing & Cropping**

```python
import os
import cv2

img = cv2.imread(os.path.join('..', 'assets', 'LauraPalmer.jpg'))
print(img.shape)  # prints the dimensions of the image (height, width, channels)
```

```python
cv2.imshow('Image', img)
cv2.waitKey(0)  # waits indefinitely until a key (any key) is pressed

cv2.destroyAllWindows() # closes all OpenCV windows, preventing memory leaks and errors
```

### **Resize**

```python
resize_img = cv2.resize(img, (300, 200))  # resize image to 300x200 pixels ,hight x width , hight = 200, width = 300(width, height but when printing shape its height, width)
print("Original image", img.shape)
print("Resized image", resize_img.shape)

cv2.imshow('Original Image', img)
cv2.imshow('Resized Image', resize_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resize to 50% of original size
# dsize=None tells OpenCV to calculate dimensions based on fx and fy
resized_img = cv2.resize(img, dsize=None, fx=0.5, fy=0.5) # dsize=None means that we are not specifying the exact size, but rather using scaling factors fx and fy to resize the image
print("Resized by 50%:", resized_img.shape)
cv2.imshow('Resized by 50%', resized_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### **Crop**

```python
img = cv2.imread(os.path.join('..', 'assets', 'LauraPalmer.jpg'))

cropped_img = img[250:600, 350:600]  # width from 200 to 500 and height from 100 to 400
print("Cropped image:", cropped_img.shape)
cv2.imshow('Cropped Image', cropped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
