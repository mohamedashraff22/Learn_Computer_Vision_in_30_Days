## The Nature of Images

* **Images as NumPy Arrays:** In OpenCV, images are treated as NumPy arrays. This allows for efficient mathematical and numerical operations using standard Python tools.
* **The `.shape` Property:** Since they are arrays, you can check an images dimensions using the `.shape` attribute.
* **Dimension Order:** The output of `.shape` follows a specific order: **(Height, Width, Number of Channels)**.

---

## Pixels and Channels

* **Pixel Composition:** Pixels are the basic building blocks of an image; they are the individual "blocks" where color or intensity information is stored.
* **BGR Color Space:** Unlike the standard RGB format, OpenCV defaults to **BGR (Blue, Green, Red)** for its three color channels.
* **Resolution Example:** A  image with 3 channels means there are 720 rows of pixels, 1280 columns, and 3 layers of color data for every single pixel.

---

## Pixel Value Ranges

* **Standard Range (8-bit):** In most cases, pixel values range from **0 to 255**.
* **Binary Images:** These images only contain two values. Pixels are either black (**0**) or white (**1** or **255**).
* **16-bit Images:** For higher precision images (2 bytes per pixel), the range expands significantly from **0 to 65,535**.

---

## Key Takeaway

To work with OpenCV, you must view an image not just as a picture, but as a **grid of numbers** where the specific value and the arrangement (Height x Width x Channel) determine what you see on the screen.
