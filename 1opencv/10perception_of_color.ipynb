# **Human Vision -> Perception of color**

# **Notes:**

### **The Big Question: Why 3 Dimensions?**

* **Observation:** Almost every popular colorspace in computer vision (RGB, BGR, HSV, LAB) utilizes exactly **3 channels** (Dimension 3).
* **The Puzzle:** The instructor explored why we need exactly three distinct values to encode "all" color information for a standard image, rather than 2, 4, or 10.

### **The Biological Answer (Trichromacy)**

* **Human Hardware:** The 3-channel standard is not a fundamental law of physics, but a result of **human biology**.
* **Cone Cells:** The human eye contains three specific types of photoreceptor cells called **Cones**.
* Since our biological "sensors" only capture three distinct signal types from the light spectrum, we only need 3 data channels (Red, Green, Blue) to recreate any color a human can perceive.



### **The "Mantis Shrimp" Thought Experiment**

* **Different Hardware:** To illustrate this, the instructor compares humans to the **Mantis Shrimp**, which has **16 types** of photoreceptor cells.
* **The Implications:** If Mantis Shrimps invented computers and OpenCV, their standard images would likely require **16 channels** to accurately represent the world as they see it. Our 3-channel images would look "black and white" or severely lacking in detail to them.

### **Key Takeaway**

Computer Vision is heavily modeled on **Human Vision**. Understanding that our digital tools (like OpenCV) are designed around human biological limitations can help you grasp why certain algorithms function the way they do.
