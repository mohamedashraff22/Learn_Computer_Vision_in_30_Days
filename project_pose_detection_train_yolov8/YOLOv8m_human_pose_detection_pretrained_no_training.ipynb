```python
import os
from google.colab import files
from ultralytics import YOLO
import shutil
```

```python
# 2. Load Model (YOLOv8 Medium Pose)

video_path = "/content/Twin Peaks summarized in one scene.mp4"

print("‚è≥ Loading YOLOv8 Medium Pose Model...")
model = YOLO('yolov8m-pose.pt')
```

```python
# 3. Run Inference
print("üöÄ Processing video... (This may take time depending on video length)")
results = model.predict(
    source=video_path,
    save=True,
    conf=0.5,      # Confidence threshold
    project="runs/pose",
    name="process_video",
    exist_ok=True
)
```

```python
# 4. Prepare Download
# Find the saved video file
output_dir = "/content/runs/pose/runs/pose/process_video/"
processed_files = [f for f in os.listdir(output_dir) if f.endswith(('.mp4', '.avi'))]

if processed_files:
    last_video = os.path.join(output_dir, processed_files[0])

    # Rename for clarity
    download_name = "pose_result.avi" # YOLO usually saves as .avi
    shutil.copy(last_video, download_name)

    print(f"‚úÖ Done! Downloading {download_name}...")
    files.download(download_name)
else:
    print("‚ùå Error: Processed video not found.")
```
